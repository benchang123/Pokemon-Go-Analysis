# Pokemon Go Analysis

## Purpose of the Analysis:

### Find the formula for CP

The purpose of this analysis is to see which statistics correlates to a Pokemon's CP the most in the game Pokemon GO. I am a lifelong Pokemon fan who often look at the statistics of the Pokemon to determine which Pokemon is best to build a strong team. When Pokemon GO was released, they introduced a new statistics called CP, which had me wondering what makes up the CP. This section will be based on inference as we are trying to see how much of the Attack, Defense, and Stamina statistics makes up the CP. 

### Prediction of the 6th Generation Pokemon's CP

I also am interested to predict the CP of the Pokemons that will be released in the near future. This will help me anticipate which Pokemons will be strong upon release. 

### Analysis of the Strongest and Weakest Pokemon Species in Pokemon GO

Upon analyzing the CP and statistics of the Pokemons, I want to see which Pokemons are currently the strongest to use in Pokemon GO.

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model as m
```

Both data from the main series game and Pokemon GO will be used in this notebook. Data was retrieved from the following sources:

Pokemon GO: https://www.kaggle.com/netzuel/pokmon-go-dataset-15-generations

Main Pokemon Games: https://gist.github.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6

```python
poke_game = pd.read_csv('https://raw.githubusercontent.com/benchang123/Pokemon-Go-Analysis/master/pokemoncomplete.csv', index_col=0, encoding= 'unicode_escape')

poke_go = pd.read_csv('https://raw.githubusercontent.com/benchang123/Pokemon-Go-Analysis/master/pogo.csv', index_col=0, encoding= 'unicode_escape')
```

## EDA

There is some duplicate data between the two datasets so a few columns were dropped and some were renamed. 

There is also faultly and wrong data in the legendary column of the Pokemon GO dataset, so I decided to use the legendary data in the main Pokemon game. There was still some wrong data in the main Pokemon game legendary column so that had to be fixed.

```python
poke_go=poke_go.reset_index()
poke_game.drop(columns=['Type 1','Type 2'], inplace=True)
poke_game.rename(columns={'Attack':'Attack_g','Defense':'Defense_g'}, inplace=True)

poke_go.drop(columns=['Legendary'], inplace=True)
poke_game.loc[poke_game.index.isin([647,648,649]),['Legendary']]=True
```

After the inital changes, the data was inspected for any NaNs and if there were some, it was addressed to so prevent it from affecting our analysis.

```python
display(poke_go.isna().sum())
poke_game.isna().sum()
```

I decided to combine the datasets to form one big table for ease of use. One caution is to make sure that all of the Pokemons make it past the merge which in this case, didn't happen.

```python
poke_go['Name']=poke_go['Name'].str.lower()
poke_game['Name']=poke_game['Name'].str.lower()

poke_go_compare=poke_go.merge(poke_game,left_on='Name',right_on='Name', how='inner')
poke_go_compare.drop(columns=['Generation_y'], inplace=True)
poke_go_compare.rename(columns={'Generation_x':'Generation'}, inplace=True)
```

From here, we are able to determine if there are any Pokemons that didn't make the final table after the merge. Most of the Pokemons in this list have multiple form.

```python
missing=poke_go.loc[poke_go['Pokedex'].isin(poke_go_compare['Pokedex'])==False,'Name']
missing
```

The problem of the missing Pokemons was addressed by mergeing the two datasets by the Pokedex number. The data was then retrieved for the missing Pokemon and then merged into our original combined dataframe to get one final dataframe with all of the Pokemons.

```python
poke_game1=poke_game.reset_index()
poke_go_missing=poke_go.merge(poke_game1,left_on='Pokedex',right_on='#', how='inner')
poke_go_missing.drop_duplicates('Name_x', keep='first', inplace=True)

poke_go_missing=poke_go_missing[poke_go_missing['Name_x'].isin(missing)]

poke_go_missing.drop(columns=['Generation_y','Name_y','#'], inplace=True)
poke_go_missing.rename(columns={'Generation_x':'Generation','Name_x':'Name'}, inplace=True)

poke_go_complete=pd.concat([poke_go_compare, poke_go_missing], axis=0)
```

## Performing Linear Regression on the Data

I chose to use multilinear regression because of its simplicity to create and as a benchmark for possible future analysis. The data does not need to be standardized since the range of each of the statistic is the same. 

I chose my loss function to be the RMSE (root means squared error). Starting with the main Pokemon game data, the HP, Attack, Defense, Special Attack, Special Defense, and Speed were chosen as the parameters to fit in the model.

```python
def rmse(y_pred,y_actual):
    return np.sqrt(np.mean((y_pred-y_actual)**2))

model=m.LinearRegression()
model.fit(poke_go_complete.loc[:,'HP':'Speed'],poke_go_complete['MaxCP'])
poke_go_complete['predCP']=model.predict(poke_go_complete.loc[:,'HP':'Speed'])
```

```python
sns.lmplot(data=poke_go_complete,x='predCP',y='MaxCP')
plt.xlabel("Predicted CP")
plt.ylabel("Actual CP")

error=round(rmse(poke_go_complete['predCP'],poke_go_complete['MaxCP']),3)
print(f'The RMSE is {error}')

print(f'The coefficient for the HP, Attack, Defense, Special Attack, Special Defense, and Speed are {np.around(model.coef_,2)} respectively.')
```

Next, a residual plot was made to see if the residual was random. The residual plot below shows that the residuals are spread apart throughout the whole x-axis, but there is a slight upward trend on the left part of the graph. For the most part, the regression seems to do a good job.

```python
residual=poke_go_complete.loc[:,['MaxCP','predCP']].diff(axis=1)
residual.drop(columns='MaxCP',inplace=True)
poke_go_complete['Game Residual']=residual.rename(columns={'predCP':'Residual'})
residual

sns.lmplot(data=poke_go_complete,x='predCP',y='Game Residual')
plt.xlabel("Predicted CP")
plt.ylabel("Residual")
```

We will do the same analysis for the Pokemon GO data. There are only three statistics that seem to make up the CP (Attack, Defense, Stamina). The Max HP is a function of the Stamina, so I did not include that in this regression.

```python
model2=m.LinearRegression()
model2.fit(poke_go_complete.loc[:,['Attack','Defense','Stamina']],poke_go_complete['MaxCP'])
poke_go_complete['pokegopred']=model2.predict(poke_go_complete.loc[:,['Attack','Defense','Stamina']])

sns.lmplot(x='pokegopred',y='MaxCP',data=poke_go_complete)
plt.xlabel("Predicted CP")
plt.ylabel("Actual CP")

error2=rmse(poke_go_complete['pokegopred'],poke_go_complete['MaxCP'])
print(f'The RMSE is {round(error2,2)}')

print(f'The coefficient for Attack, Defense, Stamina are {np.around(model2.coef_,2)} respectively.')
```

Unlike the residual plot from the Pokemon series data, this residual plot has a quadratic trend. This indicates that the multiple linear regression needs to be more complex. There are some extra features such as the primary type of the Pokemon that we can use.

```python
residual=poke_go_complete.loc[:,['MaxCP','pokegopred']].diff(axis=1)
residual.drop(columns='MaxCP',inplace=True)
poke_go_complete['Go Residual']=residual.rename(columns={'pokegopred':'Residual'})

sns.lmplot(data=poke_go_complete,x='pokegopred',y='Go Residual')
plt.xlabel("Predicted CP")
plt.ylabel("Residual")
```

To make the model more complex, the Pokemon's primary type was one hot encoded into the dataframe and then a linear regression was conducted. Adding the extra features did not seem to help the residual plot. There aren't anymore variables in the data that seem to be relevant; further analysis is needed to better fit the data, such as adding the square root or exponential of a current feature.

```python
ohc=pd.get_dummies(poke_go_complete['Primary'],drop_first=True)
poke_go_complete_ohc=pd.concat([poke_go_complete, ohc], axis=1)
interestedcolumns=poke_go_complete_ohc.iloc[:,np.r_[3,4,2,26:42]]

model3=m.LinearRegression()
model3.fit(interestedcolumns,poke_go_complete_ohc['MaxCP'])
poke_go_complete_ohc['pokegopred']=model3.predict(interestedcolumns)

sns.lmplot(x='pokegopred',y='MaxCP',data=poke_go_complete_ohc)
plt.xlabel("Predicted CP")
plt.ylabel("Actual CP")

error3=rmse(poke_go_complete_ohc['pokegopred'],poke_go_complete_ohc['MaxCP'])
print(f'The RMSE is {round(error3,2)}')
```

```python
residual=poke_go_complete_ohc.loc[:,['MaxCP','pokegopred']].diff(axis=1)
residual.drop(columns='MaxCP',inplace=True)
poke_go_complete_ohc['Go Residual']=residual.rename(columns={'pokegopred':'Residual'})

sns.lmplot(data=poke_go_complete_ohc,x='pokegopred',y='Go Residual')
plt.xlabel("Predicted CP")
plt.ylabel("Residual")
```

From both of the game data, this confirms my suspcision in the EDA that the Attack had the highest influence on the CP in the Pokemon GO data, and the same goes for the Attack and Special Attack in the main Pokemon series data. 

For fun, I wanted to see which Pokemon caused the largest residuals. From domain knowledge, it seems like pokemons with an extreme statistic is throwing off the regression (Shuckle and Bastiodon have a very high Defense and Special Defense statistic, Wobbuffet and Chansey have a very high Stamina value).

```python
# Determine Pokemon with Biggest Error

biggestDiff=poke_go_complete.loc[:,['MaxCP','predCP']].diff(axis=1).sort_values('predCP',ascending=False).head(10).index

poke_go_complete.loc[biggestDiff,'Name']
```

## Predicting the CP for the 6th Generation Pokemons in Pokemon GO

For this section, I wanted to predict the CP for the upcoming Pokemons in Pokemon GO. This can be done through using the multilinear regression model that was made in the previous section. Using the data from the main Pokemon game, we can fit a model on the current Pokemons in Pokemon GO. After that, we can use the statistics of the 6th generation Pokemon to predict the CP.

```python
gen6poke=poke_game.loc[poke_game['Generation']==6,:]

gen6poke['CP']=model.predict(gen6poke.loc[:,'HP':'Speed'])

gen6poke.sort_values('CP',ascending=False).loc[:,['Name','CP']].head(10)
```

## Determining the Strongest and Weakest Pokemon Species in Pokemon GO

In this last section, I wanted to determine the strongest and weakest species in Pokemon GO. 

I chose to standardize the data because having the highest CP might not nessecary mean that it is the best Pokemon. Since Attack is the biggest factor to CP, the Pokemon might be fragile and faint relatively quicker than another Pokemon with a slightly lower CP but a more well-rounded statistic. 

The Attack, Defense, and Stamina was standardized and used to take the dot product of the coefficients from the linear regression of the Pokemon GO data. 

```python
def standardize(data):
    return (data-np.mean(data))/np.std(data)
poke_go_complete_scaled = pd.DataFrame(standardize(poke_go_complete.loc[:,['Attack','Defense','Stamina']]))

poke_go_complete_scaled['Strength']= poke_go_complete_scaled.loc[:,'Attack':'Stamina']@model2.coef_.T
poke_go_complete_scaled['Name']=poke_go_complete['Name']
poke_go_complete_scaled['Legendary']=poke_go_complete['Legendary']

poke_go_complete_scaled=poke_go_complete_scaled[['Name','Legendary','Attack','Defense','Stamina','Strength']]
poke_go_complete_scaled.head()
```

The five strongest Pokemon from the analysis are Slaking, Regigigas, Mewtwo, Kyogre, and Groudon.

```python
poke_go_complete_scaled.sort_values("Strength", ascending=False).head()
```

I thought it might be a good idea to consider the strongest non-legendary Pokemons since they are more accessible to players and most of the strongest Pokemons are legendary Pokemons. The strongest non-Legendary Pokemons are Slaking, Garchomp, Metagross, Salamence, and Tyranitar.

```python
poke_go_complete_scaled[poke_go_complete_scaled['Legendary']==False].sort_values("Strength", ascending=False).head()
```

Next, we consider the weakest Pokemons in Pokemon GO. Magikarp, Feebas, Sunkern, Azurill, and Kricketot are the weakest Pokemons.

```python
poke_go_complete_scaled.sort_values("Strength", ascending=True).head()
```

## Conclusion

From the questions that we posed earlier, we determined that the Attack has the highest correlation to the determination of the CP. The Attack statistics nearly doubles the influence of the Defense and Stamina. 

The actual formula for CP is given by : $$CP = \frac{(Attack * \sqrt{Defense} * \sqrt{Stamina} * {CP Multiplier}^{2})}{10}$$

which shows that in the future, it would be better to conduct a non-linear regression to accurately determine the correlation of Attack, Defense, and Stamina.

We used the linear regression model to predict the CP of future Pokemon coming into Pokemon GO. Xerneas, Hoopa and Yveltal have the highest predicted CP and expect to be the most anticipated when the 6th Generation Pokemons are released.

Lastly, we determined the strongest Pokemons that are currently in the game. Slaking, Regigigas, Mewtwo, Kyogre, and Groudon are the strongest Pokemon in the game from doing a standardization of their statistics.
